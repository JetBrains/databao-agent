```
import logging

import pandas as pd

import portus

logging.basicConfig(level=logging.INFO)
```

```
data = pd.read_csv("examples/data/rfm_segmented.csv")
data
```

```
session = portus.open_session("my_session")
session.add_df(data)
```

```
# All questions are from https://medium.com/data-science-collective/testing-the-limits-of-pandasai-part-1-what-it-can-and-cant-do-to-help-data-scientists-6f0775314889

thread = session.thread()
thread.ask("How many customers are in each segment?").df()
```

```
thread.plot("light blue bar plot")
```

```
thread.plot(
    "bar plot where each segment should be a different color, and the bars should be labeled with the customer count."
)
```

```
thread.ask("What is the average monetary value for each customer segment?").df()
```

```
thread.ask("Which segment has the highest average frequency?").text()
```

```
thread.ask("Show me the top 5 customers by monetary value?").df()
```

```
thread.ask("What percentage of customers are loyal customers?").text()
```
