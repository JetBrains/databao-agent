## DataFrame

```python
import portus
s = portus.open_session("a")
import pandas as pd
s.add_df(pd.DataFrame({"k": ["re", "tr", "yt"], "num": [1,2,3]}))
pipe = s.ask("What is sum of all values?")
print(pipe.text)
```
